```python
# 4.1 Essential Concepts

# When we write "a + b" in Python, the interpreter tries to execute a.__add__(b)

# Show that this is true by trying the following:
a = 6
b = 31.5
c = 'asdfg'
a1 = a.__add__(13)
b1 = b.__add__(13)
c1 = c.__add__('13')
print(a1)
print(b1)
print(c1)

# What are the types of a, b, and c?
#a is a (int).
#b is a (float).
#c is a (str).


# Explain the results
#a1 = a.__add__(13) adds the integer 13 to the integer a (which is 6), resulting in 19.
#b1 = b.__add__(13) adds the float 13.0 to the float b (which is 31.5), resulting in 44.5.
#c1 = c.__add__('13') concatenates the string '13' to the string c (which is 'asdfg'), resulting in 'asdfg13'.


19
44.5
asdfg13

# 4.2 through 4.4 will be covered in the next assignment



# 4.5 str and repr
# How many categpries of printing are shown below for a datetime object?
# What in the Pythion code makes each one get callled?

from datetime import date
d = date(2023, 11, 7)

#print(d) a normal print statement its a very simple one where it shows the representaion of the object
#print(f'{d}') a formated print statement same as the previous in terms of how its called
#print(f'{d!s}') still a formated string but the !s is a conversion flag calls upon the string method specifically
#print(f'{d!r}') also a formated string, but the !r is a conversion flag calls the repr method specifcally
#print(str(d)) similar to the !s but this internally summons the string method
#print(repr(d)) again similar to the !r but this interally calls the repr method
#print(d.__str__()) this is also calling the string method but speciafically goes through the datetime object first
#print(d.__repr__()) almost the same as the previous but instead of calling string it calls repr



# 4.6 First-Class Objects
# What is the significance of all objects being "first-class" in Python?

#(answer) the significance of all objects being "first-class" is the fact all pieces of code are treated with the same level of urgency and importance
# as long as everything is treated equally it is unlikely that some vairables or functions go unoticed unless by human intention or error

# Explain by typing in the code to create and use the 'items' dictionary shown in section 4.6.
# Create an empty dictionary named 'items'
items = {}

items['func'] = abs

import math
items['mod'] = math

items['error'] = ValueError

nums = [1, 2, 3, 4]

items['append'] = nums.append

result_abs = items['func'](-5)

result_sqrt = items['mod'].sqrt(25)

try:
    raise items['error']('Example ValueError')
except ValueError as ve:
    print("Caught ValueError:", ve)

items['append'](5)

print(nums)

# Explain each of the lines starting with '>>>' in the section.

#items['func'](-45) retrieves the abs() function (assigned to the 'func' key) from the 'items' dictionary and applies it to the value -45. It returns the absolute value of -45, which is 45

#items['mod'].sqrt(4) retrieves the math module (assigned to the 'mod' key) from the 'items' dictionary and accesses its sqrt() method to calculate the square root of 4, resulting in 2.0

#This snippet demonstrates a try-except block where it attempts to convert the string 'a lot' to an integer (int('a lot')). However, it catches an exception using the items['error'] key (which holds the ValueError type) and prints "Couldn't convert" because converting 'a lot' to an integer raises a ValueError

#items['append'](100) accesses the append() method of the 'nums' list (assigned to the 'append' key) from the 'items' dictionary and appends the value 100 to the 'nums' list

#Prints the updated 'nums' list, which now contains the appended value 100 as the last element. The output shows the updated 'nums' list after the append() operation.



# 4.7 Using None
# Write a function that takes two parameters and returns their sum if they are
#  integers or returns None if either one is not.

# Put in some test code to show that your function works

# INSERT YOUR CODE HERE

def sum_integers(a, b):
    if isinstance(a, int) and isinstance(b, int):
        return a + b
    else:
        return None

# Test cases
print(sum_integers(5, 10))  # Output: 15
print(sum_integers(7, '8'))  # Output: None
print(sum_integers('3', 6))  # Output: None
print(sum_integers('hello', 'world'))  # Output: None


# 4.9 Object Protocol
# Create a simple object that has methods for:
#  __new__
#  __init__
#  __del__
#  __repr__
#  __str__

# Put in some test code to show how and when each of these 'magic' functions gets called.

# INSERT YOUR CODE HERE

class SimpleObject:
    def __new__(cls):
        print("__new__ method called")
        return super().__new__(cls)

    def __init__(self):
        print("__init__ method called")

    def __del__(self):
        print("__del__ method called")

    def __repr__(self):
        return "__repr__ method called"

    def __str__(self):
        return "__str__ method called"

# Test code
obj = SimpleObject()

print("Printing object using repr():", repr(obj))
print("Printing object using str():", str(obj))

del obj

```
